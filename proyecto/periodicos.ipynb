```python
import requests as rq
from bs4 import BeautifulSoup
import pandas as pd
from datetime import date

datos = {'titulo':[],
        'fecha':[],
        'texto':[],
        'periodico':[]}

portafolio = 'https://www.portafolio.co/economia'
port = rq.get(portafolio)
port_sopa = BeautifulSoup(port.text, 'lxml')
noticias = port_sopa.find_all('div', attrs={'class':'secondary-board modules'})[1].find_all('div', attrs={'class':'listing-item'})
for noticia in noticias[:-2]:
    print(noticia.a.get('href'))
    elemento = rq.get('https://www.portafolio.co'+noticia.a.get('href'))
    elemento_sopa = BeautifulSoup(elemento.text, 'lxml')
    datos['titulo'].append(elemento_sopa.find('h1', attrs={'itemprop':'headline'}).text)
    datos['fecha'].append(date.today())
    datos['texto'].append(elemento_sopa.find('div', attrs={'class':'article-content'}).text)
    datos['periodico'].append('Portafolio')

tiempo = rq.get('https://www.eltiempo.com/economia')
tiempo_sopa = BeautifulSoup(tiempo.text, 'lxml')
secciones = tiempo_sopa.find('ul', attrs={'class':'economia-menu'}).find_all('li', attrs={'itemprop':'name'})[2:]
for seccion in secciones:
    if 'https' not in seccion.a.get('href'):
        seccion_ = rq.get('https://www.eltiempo.com'+seccion.a.get('href'))
    else:
        seccion_ = rq.get(seccion.a.get('href'))
        
    seccion_sopa = BeautifulSoup(seccion_.text, 'lxml')
    noticias = seccion_sopa.find('div', attrs={'class':'notas mas-notas-bk'}).find_all('div', attrs={'class':'nota listing'})
    for noticia in noticias:
        elemento = rq.get('https://www.eltiempo.com/'+ noticia.a.get('href'))
        elemento_sopa = BeautifulSoup(elemento.text, 'lxml')
        datos['titulo'].append(elemento_sopa.find('div',attrs={'class':'titulo-principal-bk'}).text)
        datos['texto'].append(elemento_sopa.find('div', attrs={'itemprop':'articleBody'}).text)
        datos['periodico'].append('El tiempo')
        datos['fecha'].append(date.today())

    
heraldo_ = 'https://www.elheraldo.co/economia'
heraldo = rq.get(heraldo_)
heraldo_sopa = BeautifulSoup(heraldo.text, 'lxml')
noticias = heraldo_sopa.find_all('article')
for noticia in noticias[:10]:
    link = noticia.a.get('href')
    if 'https' in link:
        None
    elif 'economia' not in link:
        None
    else:
        elemento = rq.get('https://www.elheraldo.co/'+link)
        elemento_sopa = BeautifulSoup(elemento.text, 'lxml')
        datos['titulo'].append(elemento_sopa.find('article').find('h1').text)
        datos['fecha'].append(date.today())
        var = elemento_sopa.find('div', attrs={'class':'column center content-blocked'}).find_all('p')
        m = ' '.join([i.text for i in var])
        datos['texto'].append(m)
        datos['periodico'].append('El Heraldo')
        
elpais = 'https://www.elpais.com.co/economia'
elpais_ = rq.get(elpais)
elpais_sopa = BeautifulSoup(elpais_.text, 'lxml')
noticias = elpais_sopa.find('div', attrs={'class':'listing contentNews-listing latest-listing economia-listing category-listing view-list'}).find_all('div', attrs={'class':'listing-item'})
for noticia in noticias:
    if 'ads-subcategory' not in noticia.get('class'):
        elemento = rq.get('https://www.elpais.com.co'+ str(noticia.a.get('href')))
        elemento_soup = BeautifulSoup(elemento.text, 'lxml')
        datos['titulo'].append(elemento_soup.find('h1', attrs={'class':'title'}).text)
        datos['fecha'].append(date.today())
        datos['texto'].append(elemento_soup.find('div', attrs={'class':'article-content'}).text)
        datos['periodico'].append('El pais')
    else:
        print('elemento')
```

```python
datos = pd.DataFrame(datos)
#datos.to_csv(f'{date.today()}.csv')
```
